### BEGINNER LEVEL TASK

### TASK 4- Image to Pencil Sketch with Python:

```python
import cv2
```

```python
image=cv2.imread("beautiful_baby.jpeg")
```

```python
cv2.imshow("beautiful_baby",image)
cv2.waitKey(0)
```

### Converting original to greyscale 

```python
gray_image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("New_beautiful_baby",gray_image)
cv2.waitKey(0)
```

### Inverting the greyscale Image

```python
inverted_image=255-gray_image
cv2.imshow("Inverted",inverted_image)
cv2.waitKey(0)
```

### Blurring the image by using the Gaussian Function in OpenCV¶

```python
blurred=cv2.GaussianBlur(inverted_image, (21,21),0)
```

### Inverting the Blurred image andconverting to pencil sketch 

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

### looking at both the original image and the pencil sketch 

```python
cv2.imshow("original image", image)
cv2.imshow("pencil sketch", pencil_sketch)
cv2.waitKey(0)
```
